$\mathbf{Ejemplo}$ (Variable aleatoria Cauchy) Simulemos la variable aleatoria de Cauchy de parámetro $1$ que tiene por función de densidad
$$\frac{1}{\pi}\frac{1}{1+x^{2}}.$$

Ejercicio:
- Corroborar que la función así definida es una función de densidad de probabilidad
- Encontrar F_X
- Crear una gráfica en Python que muestre su forma
- Generar v.a. Cauchy a partir de una uniforme

Por el teorema anterior, para $u\in (0,1)$
$$u=\frac{1}{\pi}\arctan(x)+\frac{1}{2} \qquad \text{si y sólo si} \qquad x=\tan\left({\pi}\left(u-\frac{1}{2} \right) \right).$$

```python
import numpy as np
import plotly.graph_objects as go
```

```python
# Semilla
np.random.seed(123)
```

```python
# 1. Parámetros
n = 10000
```

 ### Simulación de la variable Cauchy a partir de una uniforme

```python
uniformes = np.random.random(n)
cauchy_simulada = np.tan(np.pi * (uniformes - 0.5))
```

### Definir la función de densidad de Cauchy

```python
def f_cauchy(x):
    return (1 / np.pi) * (1 / (1 + x**2))
```

### Generar valores teóricos para la curva

```python
x_vals = np.linspace(-10, 10, 500)
y_vals = f_cauchy(x_vals)
```

```python
#Crear gráfico interactivo con Plotly
fig = go.Figure()
```

 ### Histograma de la simulación

```python
fig.add_trace(go.Histogram(
    x=cauchy_simulada,
    nbinsx=100,
    histnorm='probability density',
    marker_color='blue',
    name='Simulación'
))
```

### Curva teórica

```python
fig.add_trace(go.Scatter(
    x=x_vals,
    y=y_vals,
    mode='lines',
    name='Densidad teórica',
    line=dict(color='red')
))

fig.update_layout(
    title='Distribución Cauchy simulada vs teórica',
    xaxis_title='x',
    yaxis_title='Densidad',
    bargap=0.05
)

fig.show()
```
